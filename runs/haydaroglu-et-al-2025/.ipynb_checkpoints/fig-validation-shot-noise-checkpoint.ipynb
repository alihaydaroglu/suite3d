```python
def calculate_noise_levels(fluor, frame_rate, method = 'pct', pct = 10):
    if method == 'pct':
        f0 = n.percentile(fluor, pct, axis=1, keepdims=True)
        df_f = (fluor - f0) / f0
    elif method == 'mean':
        f0 = n.mean(fluor,axis=1, keepdims=True)
        df_f = (fluor - f0) / f0
    elif method == 'median':
        f0 = n.median(fluor,axis=1, keepdims=True)
        df_f = (fluor - f0) / f0
    else:
        df_f = fluor

    
    noise_levels = n.nanmedian(n.abs(n.diff(df_f, axis=-1)), axis=-1) / n.sqrt(frame_rate)
    return noise_levels * 100     # scale noise levels to percent

def df_f(f,method = 'pct', pct = 10, axis=0):
    if method == 'pct':
        f0 = n.percentile(f, pct, axis=axis, keepdims=True)
        df_f = (f - f0) / f0
    elif method == 'mean':
        f0 = n.mean(f,axis=axis, keepdims=True)
        df_f = (f - f0) / f0
    elif method == 'median':
        f0 = n.median(f,axis=axis, keepdims=True)
        df_f = (f - f0) / f0
    else:
        df_f = f
    print(f0)
    return df_f
```

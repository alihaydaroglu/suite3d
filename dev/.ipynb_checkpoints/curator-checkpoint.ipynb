```python
shot_noises = qm.shot_noise_pct(fs, 3.3)
```

```python
mean_img_boxes = labeller.extract_padded_boxes(mean_img, centroids, box_size_pix)
max_img_boxes = labeller.extract_padded_boxes(max_img, centroids,box_size_pix)
vmap_boxes = labeller.extract_padded_boxes(vmap, centroids,box_size_pix)
lam_boxes = labeller.fill_lam_box(centroids, coords, lams, box_size_pix)
```

```python
mean_box3d = labeller.Box3D(mean_img_boxes)
vmap_box3d = labeller.Box3D(vmap_boxes)
lam_box3d = labeller.Box3D(lam_boxes)
```

```python
Fs_t1d = labeller.Trace1D(fs)
Fn_t1d = labeller.Trace1D(fn)
```

```python
ROIs = labeller.ROIDataset([mean_box3d, vmap_box3d, lam_box3d], [Fs_t1d, Fn_t1d])
```

```python
from suite3d.labeller import add_gaussian_noise, rotate_2d, mirror_axis, scale_intensity, shuffle_snippets
```

```python
box_transforms = {
    # 2: [lambda x : add_gaussian_noise(x, 0.5, [2]), rotate_2d],  # Mean image: add noise and rotate
    0: [rotate_2d],                     # Max image: rotate only
    # 1: [mirror_axis]  ,                  # Correlation image: mirror
    # 3: [lambda x : mirror_axis(x, 3)]                    # Correlation image: mirror
}
trace_transforms = {
    # 0: [lambda x : scale_intensity(x, None, range=(0.5, 2.0))],              # Fluorescence: scale intensity
    1: [lambda x : shuffle_snippets(x, 100, None, subset=0.5)]             # Neuropil: shuffle snippets
}
```

```python
labeller.plot_roi(ROIs[500])
```

```python
from torch.optim import Adam
```

```python
ContrastiveROIs = labeller.ContrastivePairDataset(ROIs, box_transforms, trace_transforms)
```

```python
net = labeller.CellNet(box_size_pix, nt, n_boxes = 3, n_traces = 2, latent_dim = 16, n_conv1d_layers = 2, n_conv3d_layers = 2, conv3d_kernel_sizes=[8,5],)
```

```python
net.box_encoder[0].weight.shape
```

```python
dataloader = labeller.DataLoader(ContrastiveROIs, batch_size = 32, shuffle=True)
```

```python
optimizer = Adam(net.parameters(), lr=1e-3)
```

```python
roi_dataloader = labeller.DataLoader(ROIs, batch_size = 32, shuffle=False)
```
